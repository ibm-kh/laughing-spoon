```python
import numpy as an
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI']
```

```python
ind_data  = wb.DataReader(tickers, data_source='yahoo', start='1997-1-1')['Adj Close']
```

```python
ind_data
```

###### Presentig all prices normalised by the first entry

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(20, 10))
plt.show()
```
