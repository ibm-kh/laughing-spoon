```python
import pandas as pd
from pandas_datareader import data as wb
```

```python
tickers = ['AAPL', 'MSFT', 'TSLA', 'GE', 'AMZN', 'DB']
data = pd.DataFrame()

for t in tickers:
    data[t] = wb.DataReader(t, data_source='yahoo', start='2013-01-01', end='2016-12-31')['Adj Close']
```

```python
data.head(10)
```

```python
data.plot(figsize=(20,7))
```

```python
import numpy as np
```

```python
returns = np.log(data/data.shift(1) - 1)
```
