```python
import math
import numpy as np
import numpy.random as npr
import scipy
import matplotlib.pyplot as plt
```

```python
def brownian_motion(dt=0.1,X0=0,N=10000):
    
    W = np.zeros(N+1)
    
    t = np.linspace(0, N, N+1);
    
    W[1:N + 1] = np.cumsum(scipy.random.normal(0, dt, N))
    
    return t, W
```

```python
def plot_brownian_mortion(t,W):
    plt.plot(t, W)
    plt.xlabel('Time(L)')
    plt.ylabel('Wiener-process W(t)')
    plt.title('Wiener-process')
    plt.show()
```

```python
t,W = brownian_motion(dt=0.1,X0=0,N=10000)
plot_brownian_mortion(t,W)
```
